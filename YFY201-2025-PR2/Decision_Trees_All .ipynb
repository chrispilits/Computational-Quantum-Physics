# Supervised machine Learning - Classification

## Decision Tree

In this Jupyter notebook i use all the columns to build a model with Decision Trees algorithm and compute the accuracy.

I checked and it was overfitting(train_accuracy=1 and test_accuracy=around0.6). To avoid overfitting,i changed the depth of the tree(the levels) and hence i reduced complexity. The check that with max_depth=5 overfitting does not happen is at the end of this notebook.

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\Asus\Desktop\HIGGS_8K.csv",header=None)

# Because we detected a non numerical value on the first line,we can neglect the first line.

# All rows except the first, all columns except the first
higgs = df.iloc[1:, 1:]

# All rows except the first, only the first column which is the target or label
higgs_class = df.iloc[1:, 0]
```

```python
higgs
```

```python
higgs_class
```

```python
from sklearn.model_selection import train_test_split
#test_size is the percentage of the test size to the complete dataset
# random_state is the seed. A specific pseudorandom number to split the data set
# in order to produce same splitting every time we run the script.
higgs_train, higgs_test, higgs_class_train, higgs_class_test = train_test_split(higgs, higgs_class, test_size = 0.25, random_state = 42)
```

# Decision trees classification algorithm 

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(max_depth=5,criterion = 'entropy', random_state = 0)
classifier.fit(higgs_train, higgs_class_train)
```

## The prediction of the type M or F

```python
higgs_test_pred = classifier.predict(higgs_test)
```

```python
higgs_test_pred
```

## The test set responses

```python
higgs_class_test
```

```python
#Calculating the confusion matrix and the accuracy 
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(higgs_class_test, higgs_test_pred)
print(cm)
accuracy_score(higgs_class_test, higgs_test_pred)
```

```python
# Here i check for overfitting.The train accuracy is close to test accuracy so its acceptable.

train_accuracy=accuracy_score(higgs_class_train,classifier.predict(higgs_train))
test_accuracy=accuracy_score(higgs_class_test,classifier.predict(higgs_test))

print(f"Train Accuracy: {train_accuracy:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")
```
